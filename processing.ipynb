```python
def leader_strength(row):
    return row['followers'] / (1 + row['following'])
```

```python
df_users['leader_strength'] = df_users.apply(leader_strength, axis=1)
```

```python
df_users.sort_values('leader_strength', ascending=False)[:5]
```

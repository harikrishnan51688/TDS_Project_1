```python
# Filter the DataFrame by hireable and non-hireable users
hireable_users = df_users[df_users['hireable'] == True]
non_hireable_users = df_users[df_users['hireable'] == False]

# Calculate the fraction of users with an email in each group
hireable_with_email_fraction = hireable_users['email'].notna().mean()
non_hireable_with_email_fraction = non_hireable_users['email'].notna().mean()

# Calculate the difference
fraction_difference = hireable_with_email_fraction - non_hireable_with_email_fraction
print(f"Difference in fraction of users with email (hireable - non-hireable): {fraction_difference:.3f}")
```

```python
# Filter out users with missing names
df_users_with_names = df_users[df_users['name'].notna()]

# Extract the last word in each name, assuming it's the surname
df_users_with_names['surname'] = df_users_with_names['name'].str.strip().str.split().str[-1]

# Count occurrences of each surname
surname_counts = df_users_with_names['surname'].value_counts()

# Find the maximum count
max_count = surname_counts.max()

# Get the most common surnames (those with the maximum count)
most_common_surnames = surname_counts[surname_counts == max_count].index.sort_values().tolist()
most_common_surnames_str = ", ".join(most_common_surnames)

# Output the result
print(f"Most common surnambe(s): {most_common_surnames_str}")
print(f"Number of users with the most common surname: {max_count}")
```
